```python
path = "datasets/Amharic character Image dataset/models/unet2-arct/best_val_loss_model.h5"

model = load_model(path)
```

```python


# External Images
# img = cv2.imread(r'_cotet/test-images/fidel (15).png') # path to image file


# Dataset .png test images
img = cv2.imread(r'../DATASET/Amharic character Image dataset/Type-I/image_80/010999.bin.png') # path to image file



img_copy = img.copy()
# print(img_copy.shape)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # converted from BGR to RGB
img = cv2.resize(img, (400,440)) # resize the image to our required dimensions
# print(img.shape)
# img_copy =cv2.GaussianBlur(img_copy, (7,7), 0)
# print(img_copy.shape)
img_gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY) # to greyscale\
# print(img_gray.shape)
# _, img_thresh = cv2.threshold(img_gray, 100, 255, cv2.THRESH_BINARY_INV) # thresholding because keeps the image smooth without any sort of haz gray colors in the image that could lead to wring predictions
# print(img_thresh.shape)
img_final = cv2.resize(img_gray, (32,32)) # resized into the dimensions that the model takes as input
# print(img_final.shape)
img_final = np.reshape(img_final, (1, 32, 32, 1)) # reshaping so that it can be used as model input
```

```python
img_pred = AMHARIC_ALPHABET[np.argmax(model.predict(img_final))]

cv2.putText(img, "Image: ", (20,25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color = (255,0,0))


fontpath = "datasets/Amharic character Image dataset/nyala.ttf"
font = ImageFont.truetype(fontpath, 32)
img_pil = Image.fromarray(img)
draw = ImageDraw.Draw(img_pil)
draw.text((20, 410), "Prediction: " + img_pred, font=font, fill=(255,0,30,0))
img = np.array(img_pil)


# for the English prediciton
# cv2.putText(img, "Prediction: " + img_pred, (20,410), cv2.FONT_HERSHEY_DUPLEX, 1.3, color = (255,0,30),)

cv2.imshow('External Predicition', img)


img_pil = Image.fromarray(img)


while (1):
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
```
